```python
import re
import pandas as pd
import bs4
import requests
import spacy
from spacy import displacy


nlp = spacy.load("pl_core_news_sm")
from spacy.matcher import Matcher 
from spacy.tokens import Span 

import networkx as nx

import matplotlib.pyplot as plt
from tqdm import tqdm

pd.set_option('display.max_colwidth', 200)
%matplotlib inline
```

```python
candidate_sentences = pd.read_csv("sents.csv")
candidate_sentences.shape
```

```python
persName = {}
placeName = {}
orgName = {}
geogName = {}

for i in range(len(candidate_sentences) - 1):
  sent = list(candidate_sentences.iloc[i])[0]
  doc = nlp(sent)
  for named_entity in doc.ents:
    if named_entity.label_ == "persName":
        named_entity = str(named_entity)
        if named_entity in persName:
          persName[named_entity] = persName[named_entity] + 1
        else:
          persName[named_entity] = 1

    elif named_entity.label_ == "placeName":
        named_entity = str(named_entity)
        if named_entity in placeName:
          placeName[named_entity] = placeName[named_entity] + 1
        else:
          placeName[named_entity] = 1
    elif named_entity.label_ == "orgName":
        named_entity = str(named_entity)
        if named_entity in orgName:
          orgName[named_entity] = orgName[named_entity] + 1
        else:
          orgName[named_entity] = 1

    elif named_entity.label_ == "geogName":
        named_entity = str(named_entity)
        if named_entity in geogName:
          geogName[named_entity] = geogName[named_entity] + 1
        else:
          geogName[named_entity] = 1

```

```python
persName = dict(sorted(persName.items(), key=lambda item: item[1], reverse=True))
placeName = dict(sorted(placeName.items(), key=lambda item: item[1], reverse=True))
orgName = dict(sorted(orgName.items(), key=lambda item: item[1], reverse=True))
geogName = dict(sorted(geogName.items(), key=lambda item: item[1], reverse=True))
```

```python
# load csv module
import csv

def dict_to_csv(dic, name):
  # open file for writing, "w" is writing
  w = csv.writer(open(name + ".csv", "w"))

  # loop over dictionary keys and values
  for key, val in dic.items():

      # write every key and value to file
      w.writerow([key, val])
```

```python
dict_to_csv(persName, "persName")
```

```python
dict_to_csv(orgName, "orgName")
dict_to_csv(geogName, "geogName")
dict_to_csv(placeName, "placeName")
```
